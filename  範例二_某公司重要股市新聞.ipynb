# 範例二 : 某公司重要股市新聞(外)

```python
import requests
import bs4
import pandas as pd
from bs4 import BeautifulSoup
import re
import  urllib.parse  


keyword='台積電' # 要查的公司，改這裡

url = 'https://tw.finance.yahoo.com/news_search.html?ei=Big5&q=' + urllib.parse.quote(keyword.encode('big5')) # 要抓取的網址

dict = {"標題":[],"內容": []}

list_req = requests.get(url)#請求網站
list_req.encoding = 'big5' #F12 Content-Type
soup = BeautifulSoup(list_req.text,'html.parser')#將整個網站的程式碼爬下來

getAllNew= soup.find('table',{'id':'newListContainer'})#抓到新聞的table

print("連結網址 :",url)
if isinstance(getAllNew,bs4.element.Tag):
    getNews= getAllNew.find_all('td',{'valign':'top'})

    for i in range (0,len(getNews)-1,2):
        dict['標題'].append(getNews[i].text)
        dict['內容'].append(getNews[i+1].text)
        
df = pd.DataFrame(dict)
df.to_csv('./Data/{0}_重要股市新聞.csv'.format(keyword), encoding='utf_8_sig')                                         # 儲存成 csv 檔

df
```
